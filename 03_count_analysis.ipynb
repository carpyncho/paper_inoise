```python
import pickle
from collections import Counter

import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
with open("./data/count.pkl") as fp:
    count = pickle.load(fp)
with open("./data/count_high.pkl") as fp:
    counth = pickle.load(fp)
with open("./data/count_low.pkl") as fp:
    countl = pickle.load(fp)
```

```python
total = Counter()
map(total.update, count);
total.pop(0)

high = Counter()
map(high.update, counth)
# high.pop(0)

low = Counter()
map(low.update, countl)
# low.pop(0)

del count, counth, countl
```

```python
from libs import cnt_stats
```

```python
cnt_stats.mean(total), cnt_stats.mean(high), cnt_stats.mean(low)
```

```python
plt.plot(total.values())
plt.axvline(64, color="r", ls="--", alpha=.5, label="Pico 64")
plt.axvline(130, color="g", ls="--", alpha=.5,  label="Pico 130")
plt.axvline(cnt_stats.mean(total), alpha=.5, color="k", ls="--", label="Mean {}".format(cnt_stats.mean(total)))
plt.xlabel("Conteo")
plt.ylabel("Frecuencia")
plt.legend()
```
